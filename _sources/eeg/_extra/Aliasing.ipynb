```python
import pylab
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['savefig.dpi'] = 300
%matplotlib inline
```

```python
srate = 250
time = 1 # sec
freq = 10 # Hz

x = np.linspace(0, time, srate * time)

y = np.sin(2 * np.pi * freq * x)

plt.plot(x, y, color='.67')

# plt.savefig("aliasing_10Hz_at_500samp.svg")
plt.show()
```

```python
fig, axs = plt.subplots(2, 3, figsize=[15, 6])


for i, srate2 in enumerate(range(3, 21, 3)):

    steps2 = round((srate * time) / (srate2 * time))

    x2 = x[::steps2]
    y2 = y[::steps2]

    axs.flatten()[i].plot(x, y, color='k', linestyle=':')
    axs.flatten()[i].plot(x2, y2, marker='o', color='red')
    axs.flatten()[i].set_axis_off()
    # plt.title(str(srate2) + ' Hz')
    
plt.savefig('aliasing_subplots.png')
plt.show()
```

```python
for srate2 in range(3, 20, 1):

    steps2 = round((srate * time) / (srate2 * time))

    x2 = x[::steps2]
    y2 = y[::steps2]

    plt.plot(x, y, color='k', linestyle=':')
    plt.plot(x2, y2, marker='o', color='red')
    plt.title(str(srate2) + ' Hz')
    plt.savefig('aliasing_srate_' + str(srate2) + '.png')
    plt.show()
```
